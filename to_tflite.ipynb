```python
from ultralytics import YOLO
import os

MODEL_PATH = "/content/drive/MyDrive/best.pt"
# √áƒ±ktƒ± TFLite dosyasƒ±nƒ±n adƒ±
TFLITE_MODEL_NAME = "disordo_best.tflite"
# Modelin beklediƒüi giri≈ü boyutu
IMAGE_SIZE = 640

try:
    model = YOLO(MODEL_PATH)
    print("Orijinal model (.pt) ba≈üarƒ±yla y√ºklendi!")
except Exception as e:
    print(f"{MODEL_PATH}")
    raise

# Modeli TFLite formatƒ±na d√∂n√º≈üt√ºr ve kaydet
try:
    print(f"\nModel {TFLITE_MODEL_NAME} olarak TFLite formatƒ±na d√∂n√º≈üt√ºr√ºl√ºyor")

    tflite_path = model.export(
        format="tflite",
        imgsz=IMAGE_SIZE,
        name="disord_best"  # √áƒ±ktƒ± klas√∂r√º/dosya adƒ± i√ßin 'name' kullanƒ±ldƒ±
    )

    print("----------------------------------------------------")

    print(f"TFLite dosya yolu: {tflite_path}")
    print("----------------------------------------------------")

except Exception as e:
    print(f"\n Hata: TFLite d√∂n√º≈ü√ºm√º sƒ±rasƒ±nda bir sorun olu≈ütu.")
    print(e)

!ls -lh runs/export/disord_best/disord_best.tflite
```
